Новый набор данных

```python
form1 = pd.read_excel('Горные породы. Данные.xlsx', 'Петрография')
```

```python
form1[['temp']] = pd.DataFrame(form1['Координаты'].str.split(',').str.join(''))

new_form = pd.DataFrame(form1['temp'].str.split("N ",n = 2, expand = True))
new_form = pd.DataFrame(new_form[1].str.split("°",n = 2, expand = True))
form1[['g']] = pd.DataFrame(new_form[0])
new_form = pd.DataFrame(new_form[1].str.split("'",n = 2, expand = True))
form1[['min']] = pd.DataFrame(new_form[0])
form1[['sec']] = pd.DataFrame(new_form[1].replace(".", ","))

form1['x'] = form1['g'].astype('float')+form1['min'].astype('float')/60+form1['sec'].astype('float')/3600

new_form = pd.DataFrame(form1['temp'].str.split("E ",n = 2, expand = True))
new_form = pd.DataFrame(new_form[1].str.split("°",n = 2, expand = True))
form1[['g']] = pd.DataFrame(new_form[0])
new_form = pd.DataFrame(new_form[1].str.split("'",n = 2, expand = True))
form1[['min']] = pd.DataFrame(new_form[0])
form1[['sec']] = pd.DataFrame(new_form[1])

form1['y'] = form1['g'].astype('float')+form1['min'].astype('float')/60+form1['sec'].astype('float')/3600

form1.drop(['temp'],axis=1, inplace=True)
form1.drop(['g'],axis=1, inplace=True)
form1.drop(['min'],axis=1, inplace=True)
form1.drop(['sec'],axis=1, inplace=True)
```
